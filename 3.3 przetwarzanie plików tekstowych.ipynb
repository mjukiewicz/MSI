```python
file_in = open("spam.csv", "r")
file_out = open("spam_marcin.csv", "w")
for line in file_in:
    line = line.replace("v1,v2","v1\tv2")
    line = line.replace("ham,","ham\t")
    line = line.replace("spam,","spam\t")
    line = line.replace(",,,","")
    file_out.write(line)
```
