```python
import pandas as pd

df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"])
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)

model = kmeans.fit(df.iloc[:,2:])

labels = model.predict(df.iloc[:,2:])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
kmeans_7 = KMeans(n_clusters=7)

model_7 = kmeans_7.fit(df.iloc[:,2:])

labels_7 = model_7.predict(df.iloc[:,2:])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels_7)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```
